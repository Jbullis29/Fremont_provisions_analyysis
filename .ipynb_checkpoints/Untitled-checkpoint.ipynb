```python
import pandas as pd
import numpy as np
```

```python
service=[]
food=[]
environment=[]
experience=[]
options=[]
reccomend=[]
total=[]

while True:
    service.append(input('how would you rate the service:'))
    food.append(input('how would you rate the food:'))
    environment.append(input('How would you rate overall environment:'))
    experience.append(input('How would you rate the overall experience:'))
    options.append(input('How would you rate menu options:'))
    reccomend.append(input('Would you reccomend us to family and friends:'))
    total.append(input('Was the total what you expect it to be: 1 for yes, 2 for no lower and 3 for no higher'))
    print('----------------------------------------------------------------')
    if input('would you like to add more (y for yes, n for no):') == 'n':
        break
```

```python
fp_analysis = pd.DataFrame({'service':service,
                           'food':food,
                           'environment':environment,
                           'experience':experience,
                           'menu_option':options,
                           'reccomed_us': reccomend,
                           'total_match_expectations':total})
```
